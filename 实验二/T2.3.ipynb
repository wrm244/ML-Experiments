```python
# 导入必要的库
import torch
import torchvision
import matplotlib.pyplot as plt

# 下载 mnist 数据集
train_data = torchvision.datasets.MNIST(root='data', train=True, transform=torchvision.transforms.ToTensor(), download=True)
test_data = torchvision.datasets.MNIST(root='data', train=False, transform=torchvision.transforms.ToTensor())

# 定义 DataLoader 对象
train_loader = torch.utils.data.DataLoader(train_data, batch_size=100, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=100, shuffle=True)

# 定义 CNN 模型的类
class CNN(torch.nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # 定义两个卷积层
        self.conv1 = torch.nn.Conv2d(1, 16, 3) # 输入通道为 1，输出通道为 16，卷积核大小为 3x3
        self.conv2 = torch.nn.Conv2d(16, 32, 3) # 输入通道为 16，输出通道为 32，卷积核大小为 3x3
        # 定义两个池化层
        self.pool = torch.nn.MaxPool2d(2) # 池化核大小为 2x2
        # 定义两个全连接层
        self.fc1 = torch.nn.Linear(32 * 5 * 5, 800) # 输入维度为 32 * 5 * 5（经过两次卷积和池化后的特征图大小），输出维度为 800
        self.fc2 = torch.nn.Linear(800, 5) # 输入维度为 800，输出维度为 5（类别数）
        # 定义一个 softmax 层
        self.softmax = torch.nn.Softmax(dim=1) # 在第一个维度上进行 softmax

    def forward(self, x):
        # 前向传播过程
        x = self.pool(torch.nn.functional.relu(self.conv1(x))) # 第一次卷积
        x = self.pool(torch.nn.functional.relu(self.conv2(x))) # 第二次卷积
        x = x.view(-1, 32 * 5 * 5) # 将特征图展平为一维向量
        x = torch.nn.functional.relu(self.fc1(x)) # 第一个全连接层
        x = self.fc2(x) # 第二个全连接层
        x = self.softmax(x) # softmax 层
        return x

# 创建一个 CNN 模型的实例
model = CNN()

# 定义一个损失函数
criterion = torch.nn.CrossEntropyLoss()

# 定义一个优化器
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# 定义两个列表，用于存储每个周期的准确率
train_acc = []
test_acc = []

# 定义一个训练函数
def train(epoch):
    # 设置模型为训练模式
    model.train()
    # 初始化训练损失和正确预测的数量
    train_loss = 0
    correct = 0
    # 遍历训练数据
    for batch_idx, (data, target) in enumerate(train_loader):
        # 将数据和标签转换为适合模型的格式
        data = data # 数据已经是张量格式，无需转换
        target = target // 2 # 将标签为 0 和 1，2 和 3，4 和 5，6 和 7，8 和 9 的数字分别分为一类（共 5 类）
        # 将数据和标签放到计算设备上，如 CPU 或 GPU
        data = data.to(device)
        target = target.to(device)
        # 清空之前的梯度
        optimizer.zero_grad()
        # 前向传播，得到预测结果
        output = model(data)
        # 计算损失
        loss = criterion(output, target)
        train_loss += criterion(output, target).item()
        # 得到预测的类别，即概率最大的那一类
        pred = output.argmax(dim=1, keepdim=True)
        # 累加正确预测的数量
        correct += pred.eq(target.view_as(pred)).sum().item()
        # 反向传播，计算梯度
        loss.backward()
        # 更新参数
        optimizer.step()
        # 打印训练进度和损失信息
        if batch_idx % 100 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
    # 将训练准确率添加到列表中
    train_acc.append(100. * correct / len(train_loader.dataset))

# 定义一个测试函数，添加一个参数 epoch
def test(epoch):
    # 设置模型为评估模式
    model.eval()
    # 初始化测试损失和正确预测的数量
    test_loss = 0
    correct = 0
    # 遍历测试数据
    with torch.no_grad(): # 不计算梯度，节省内存和时间
        for data, target in test_loader:
            # 将数据和标签转换为适合模型的格式
            data = data # 数据已经是张量格式，无需转换
            target = target // 2 # 将标签为 0 和 1，2 和 3，4 和 5，6 和 7，8 和 9 的数字分别分为一类（共 5 类）
            # 将数据和标签放到计算设备上，如 CPU 或 GPU
            data = data.to(device)
            target = target.to(device)
            # 前向传播，得到预测结果
            output = model(data)
            # 累加测试损失
            test_loss += criterion(output, target).item()
            # 得到预测的类别，即概率最大的那一类
            pred = output.argmax(dim=1, keepdim=True)
            # 累加正确预测的数量
            correct += pred.eq(target.view_as(pred)).sum().item()

    # 计算并打印平均测试损失和准确率信息
    test_loss /= len(test_loader.dataset)
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    # 将测试准确率添加到列表中
    test_acc.append(100. * correct / len(test_loader.dataset))
    
    # 如果是第 1、5 或 10 个周期，则进行直观验证
    if epoch in [1, 5, 20 , 30 ,50]:
        # 遍历测试数据的前 10 个样本
        for i in range(10):
            # 获取数据和标签，并分开存储
            data, target = test_data[i] 
            # 将数据转换为适合模型的格式，并增加一个批次维度
            data = data.unsqueeze(0) 
            # 将数据放到计算设备上，如 CPU 或 GPU
            data = data.to(device)
            # 前向传播，得到预测结果
            output = model(data)
            # 得到预测的类别，即概率最大的那一类
            pred = output.argmax(dim=1, keepdim=True).item()
            # 绘制图像和预测结果
            plt.subplot(2, 5, i + 1)
            plt.imshow(data.squeeze().cpu().numpy(), cmap='gray')
            plt.title('Pred: {}'.format(pred))
        plt.suptitle('Epoch: {}'.format(epoch))
        plt.show()

# 定义一个主函数
def main():
    # 定义计算设备，如 CPU 或 GPU
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    # 将模型放到计算设备上
    model.to(device)
    # 定义训练和测试的周期数
    epochs = 50
    # 循环进行训练和测试
    for epoch in range(1, epochs + 1):
        train(epoch)
        test(epoch)
    # 使用 matplotlib.pyplot.plot 绘制两个列表的折线图，并添加标题，标签和图例
    plt.plot(train_acc, label='Train accuracy')
    plt.plot(test_acc, label='Test accuracy')
    plt.title('Accuracy of CNN on MNIST dataset')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.show()

# 检查当前模块是否为主模块
if __name__ == '__main__':
    # 调用主函数
    main()   
```
